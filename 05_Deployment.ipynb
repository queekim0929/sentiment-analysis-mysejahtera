Multinomial Naive Bayes has been choosen as our model for our MySejahtera Sentiment Analysis app.
In the deployment stage, we will deploy our model to an Web API. To do this, we have choose Gradio as our deployment library. Gradio is a free and open-source Python library that allows us to develop an easy-to-use customizable component demo for our machine learning model that anyone can use anywhere.

This Web API will prompt an input from user. User are allow to key in their review just like giving review on the google play store. Once the submit button is clicked, the sentence/input will send to our model. After the model predict the result for the input, the result will be shown at output tab by telling the users that "Positive Sentiment Detected" or Negative Sentiment Detected".

Gradio is a free library and it allow us to generate a public URL that last for 72 hours. In future, to host the Web API permanently, 'Hugging Face'(a community and data science platform that provides hosting for Gradio) can be consider. It has various of pricing starts from 0.6usd. ## Deployment

```python
# load the model file (Note:Directory must be same for notebook and model file)
from joblib import load
clf = load('mnb_model.joblib')

import gradio as gr

# Function that will take the user input send to the model for analysis and return a result
def prediction(user_input):
    #user_input.encode('unicode_escape')
    input1=[user_input]
    if clf.predict(input1) == 0:
        return("Positive Sentiment Detected")
    elif clf.predict(input1) == 1 :
        return ("Negative Sentiment Detected")
    else:
        return ("Error")
        

    
# Gradio interface setup
demo = gr.Interface(fn=prediction, inputs="text", outputs="text", title="User Sentiment Prediction", 
                    description="Please type a review for the app.", allow_flagging="never")

# launch the Web API and enable public URL sharing
demo.launch(share=True)  
```
